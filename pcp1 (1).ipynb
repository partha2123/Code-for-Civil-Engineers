# 3.0 Code

## 3.1 Importing libraries and generating random numbers

```python
import numpy as np
import scipy.stats as st

# random variates to generate random numbers
Nsim = 440000
x1= st.norm.rvs(loc=0,scale=1,size=Nsim) #random number generation for values of x1
x2= st.norm.rvs(loc=0,scale=1,size=Nsim) #random number generation for values of x2
#The location (loc) keyword specifies the mean and the scale (scale) keyword specifies the standard deviation

print ('Values of x1:', x1)
print ('Values of x2:', x2)
```

## 3.2 Defining the given function and looking at the results

```python
gx= lambda x1,x2: 2-(x1+x2)/np.sqrt(2) #A lambda function is used to generate given problem

# A lambda function can take any number of arguments, but can only have one expression

idx= np.arange(0,Nsim,1) #the array creation routines based on numerical ranges
res=[]

for i in idx:
    gg=gx(x1[i],x2[i])
    res.append(gg)  # adding the values of gg to res
res
```

## 3.3 Calculating failure probability

```python
a=list(res)
pos_count,neg_count=0,0

for num in a: #the loop for counting positive and negative number
    if num>=0:
        pos_count+=1
    else:
        neg_count+=1
        
print("Negative numbers in the list: ", neg_count)

print("positive numbers in the list: ", pos_count)


pof=neg_count/Nsim    #formula for failure probability

print("Failure probability: ", pof)
```

## 3.4 Plotting the graph

```python
import numpy as np
import matplotlib.pyplot as plt


  
  

# sort the data in ascending order
x = np.sort(res)
  
# get the cdf values of y
y = np.arange(Nsim) / float(Nsim)
  
# plotting
plt.xlabel('Results')
plt.ylabel('Probability')
plt.grid()  
plt.title('CDF using sorting the data')
  
plt.plot(x, y, marker='o')
```
