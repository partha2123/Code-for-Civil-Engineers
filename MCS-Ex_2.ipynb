# 3.0 Code

## 3.1 Importing libraries and generating random numbers

```python
import numpy as np
import scipy.stats as st
Nsim =1720000
x1= st.norm.rvs(loc=10,scale=5,size=Nsim) 
x2= st.norm.rvs(loc=9.9,scale=5,size=Nsim)

print ('Values of x1:', x1)
print ('Values of x2:', x2)
```

## 3.2 Defining the function and looking at the results

```python
gx= lambda x1,x2: ((x1)**3 + (x1)**2 * (x2) + (x2)**3- 18)

idx= np.arange(0,Nsim,1)
res=[]

for i in idx:
    gg=gx(x1[i],x2[i])
    res.append(gg)
res
```

## 3.3 Calculating failure probability of  equation

```python
a=list(res)
pos_count,neg_count=0,0
j=0
for j in range(len(res)):
    if res[j]>=0:
        pos_count+=1
    else:
        neg_count+=1

print("Negative numbers in the list: ", neg_count)

print("positive numbers in the list: ", pos_count)


pof = neg_count/Nsim
print("Failure probability: ", pof)
```

## 3.4 Plotting the graph

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd


# sort the data in ascending order
x = np.sort(res)
  
# get the cdf values of y
y = np.arange(Nsim) / float(Nsim)
  
# plotting
plt.xlabel('results')
plt.ylabel('probability')
  
plt.title('CDF using sorting the data')
  
plt.plot(x, y, marker='o')
```
