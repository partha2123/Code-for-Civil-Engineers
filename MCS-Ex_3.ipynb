# 3.0 Code

## 3.1 Importing libraries and generating random numbers




```python
import numpy as np
import scipy.stats as st
import math

Nsim1 = 3950000
x1= st.norm.rvs(loc=0,scale=1,size=Nsim1) 
x2= st.norm.rvs(loc=0,scale=1,size=Nsim1)

print ('Values of x1:', x1)
print ('Values of x2:', x2)
```

## 3.2 Defining the first function and looking at the results

```python
g1x= lambda x1,x2: 3-1-x2 + math.exp (((-x1*2)/10)+(x1/5)*4) 

idx= np.arange(0,Nsim1,1)
res1=[]

for i in idx:
    gg1=g1x(x1[i],x2[i])
    res1.append(gg1)
res1
```

## 3.3 Calculating failure probability of first equation



```python
a=list(res1)
pos_count1,neg_count1=0,0

for num in a:
    if num>=0:
        pos_count1+=1
    else:
        neg_count1+=1
        
print("Negative numbers in the list: ", neg_count1)

pof1 = neg_count1/Nsim1
print("Failure probability using first equation: ", pof1)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

  

# sort the data in ascending order
x = np.sort(res1)
  
# get the cdf values of y
y = np.arange(Nsim1) / float(Nsim1)
  
# plotting
plt.xlabel('results')
plt.ylabel('probability')
  
plt.title('CDF using sorting the data')
  
plt.plot(x, y, marker='o')
```

## 3.4 Defining the second function and looking at the results

```python
g2x= lambda x1, x2: (4.5) - (x1*x2) 
Nsim2 = 160000
idx2= np.arange(0,Nsim2,1)
res2=[]

for i in idx2:
    gg2=g2x(x1[i],x2[i])
    res2.append(gg2)
res2
```

# 3.5 Plotting the graph

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

  

# sort the data in ascending order
x = np.sort(res2)
  
# get the cdf values of y
y = np.arange(Nsim2) / float(Nsim2)
  
# plotting
plt.xlabel('results')
plt.ylabel('probability')
  
plt.title('CDF using sorting the data')
  
plt.plot(x, y, marker='o')
```

## 3.5 Calculating failure probability of second  equation and minimum failure probability from the results obtained

```python
a=list(res2)
pos_count2,neg_count2=0,0

for num in a:
    if num>=0:
        pos_count2+=1
    else:
        neg_count2+=1
        
print("Negative numbers in the list: ", neg_count2)

pof2 = neg_count2/Nsim2
print("Failure probability using second equation: ", pof2)

pof=min(pof1,pof2)
print("Minimum Failure probability : ", pof)
```
